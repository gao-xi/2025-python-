```python
%%time
data = pd.read_csv(r'C:\Users\冤种千斤\Desktop\代码\第二次作业\000001.csv')#定义data
data['Day'] = pd.to_datetime(data['Day'],format='%Y/%m/%d')
data.set_index('Day', inplace = True)
data.sort_values(by = ['Day'],axis=0, ascending=True)
data#整理表格
```

```python
from datetime import date


data_new = date ['1995-01':'2025-08'].copy()
data_new['Close'] = pd.to_numeric(data_new['Close'])
data_new['Preclose'] = pd.to_numeric(data_new['Preclose'])
data_new#在定义data后（新起一行就出现问题）准备data_new
```

```python
#向量化计算
data_new["Raw_return"] = data_new["Close"] / data_new["Preclose"] - 1
data_new["Log_return"] = np.log(data_new["Close"] / data_new["Preclose"])
#使用函数计算
data_new["Pct_change_return"] = data_new["Close"].pct_change()
#使用diff和div组合
data_new["Diff_div_return"] = data_new["Close"].diff() / data_new["Close"].shift(1)
#列表格看差异
data_new
```

```python
# 大数据集上极慢，仅为教学演示
if 'Loop_return' not in data_new.columns:
    data_new['Loop_return'] = np.nan
for i in range(len(data_new)):
    close = data_new.iloc[i]['Close']
    preclose = data_new.iloc[i]['Preclose']
    data_new.loc[data_new.index[i], 'Loop_return'] = close / preclose - 1

#使用zip+循环（更Pythonic的遍历方式）
close_vals = data_new['Close'].values
preclose_vals = data_new['Preclose'].values
loop_return_vals = []
for close, preclose in zip(close_vals, preclose_vals):
    if preclose != 0 and not np.isnan(preclose):
        loop_return_vals.append(close / preclose - 1)
    else:
        loop_return_vals.append(np.nan)
data_new['Loop_return2'] = loop_return_vals

# numpy向量化操作（高效简洁，推荐）
data_new['Numpy_return'] = (data_new['Close'].values / data_new['Preclose'].values) - 1

# 展示结果
data_new
```

```python
# 计算月度收益
# 按月末（'M'）重采样，对对数收益率求和(我的数据用ME报错）难受
Month_data1 = data_new.resample('M')['Log_return'].sum().to_frame(name='Log_return')

# 将对数收益率转换为原始收益率（公式：原始收益率 = exp(对数收益率总和) - 1）
Month_data1['Raw_Return'] = np.exp(Month_data1['Log_return']) - 1

#添加年月列，便于按时间维度分析
Month_data1['Year'] = Month_data1.index.year
Month_data1['Month'] = Month_data1.index.month
# 展示结果（查看前几行）
Month_data1.head()
```

```python
# 方法2：基于月末价格计算月度收益率（贴合金融实务，直接用月末价格推导）
# 步骤1：按月末（'M'）重采样，提取每月最后一个交易日的收盘价
monthly_data2 = data_new.resample('M')['Close'].last().to_frame(name='Close')

# 步骤2：计算上月末收盘价（作为本月的Preclose）
monthly_data2['Preclose'] = monthly_data2['Close'].shift(1)

# 步骤3：计算月度原始收益率和对数收益率
monthly_data2['Raw_return'] = monthly_data2['Close'] / monthly_data2['Preclose'] - 1
monthly_data2['Log_return'] = np.log(monthly_data2['Close'] / monthly_data2['Preclose'])

# 步骤4：添加年月维度，便于时间序列分析
monthly_data2['Year'] = monthly_data2.index.year
monthly_data2['Month'] = monthly_data2.index.month

# look一下
monthly_data2.head()
```

```python
# “1990-12-12”日期格式 里面的year年份 month月份 day 直接提出取来
data_new2 = data_new.copy()
data_new2['year'] = data_new2.index.year
data_new2['month'] = data_new2.index.month
data_new2
```

```python
# 首先提取年月信息
data_new3 = data_new.copy()
data_new3['year'] = data_new3.index.year
data_new3['month'] = data_new3.index.month

# 使用groupby按年月分组，然后对每组的对数收益率求和
Month_data3 = data_new3.groupby(['year', 'month'])['Log_return'].sum().to_frame()
Month_data3['Raw_Return'] = np.exp(Month_data3['Log_return']) - 1

# 显示结果
Month_data3
```

```python
# 方法4：分组后灵活计算月度对数收益率（简化逻辑，提升效率）
monthly_data4 = data_new3.groupby(['year', 'month'])['Log_return'].sum().to_frame(name='Log_return')
monthly_data4['Raw_Return'] = np.exp(monthly_data4['Log_return']) - 1  # 对数→原始收益率转换

# 方法5：分组后同时计算多个统计量（整理列名，提升可读性）
monthly_data5 = data_new3.groupby(['year', 'month']).agg(
    Log_return_sum=('Log_return', 'sum'),
    Log_return_mean=('Log_return', 'mean'),
    Log_return_std=('Log_return', 'std'),
    Log_return_count=('Log_return', 'count'),
    Raw_return_mean=('Raw_return', 'mean'),
    Raw_return_std=('Raw_return', 'std')
)

# 展示结果（Jupyter中display更美观）
print("方法4结果：")
display(monthly_data4.head())

print("\n方法5结果（多统计量聚合）：")
display(monthly_data5.head())
```

```python
#计算季度收益
#使用QE参数（依旧不让用qe）
Quarter_data1 = data_new.resample('Q')['Log_return'].sum().to_frame(name='Log_return')
Quarter_data1['Raw_Return'] = np.exp(Quarter_data1['Log_return']) - 1
Quarter_data1['Year'] = Quarter_data1.index.year
Quarter_data1['Quarter'] = Quarter_data1.index.quarter
#使用季度价格
Quarter_data2 = data_new.resample('Q')['Close'].last().to_frame()
Quarter_data2['Preclose'] = Quarter_data2['Close'].shift(1)
Quarter_data2['Raw_return'] = Quarter_data2['Close'] / Quarter_data2['Preclose'] - 1
Quarter_data2['Log_return'] = np.log(Quarter_data2['Close']) - np.log(Quarter_data2['Preclose'])

# 显示结果
print("季度对数收益率汇总:")
Quarter_data1
print("\n季度末价格计算的收益率:")
Quarter_data2
```

```python
# 计算年度收益率
# 方法1：使用resample函数的'YE'参数（年末）
Year_data1 = data_new.resample('Y')['Log_return'].sum().to_frame(name='Log_return')
Year_data1['Raw_Return'] = np.exp(Year_data1['Log_return']) - 1

# 方法2：使用年末价格计算
Year_data2 = data_new.resample('Y')['Close'].last().to_frame()
Year_data2['Preclose'] = Year_data2['Close'].shift(1)
Year_data2['Raw_return'] = Year_data2['Close'] / Year_data2['Preclose'] - 1
Year_data2['Log_return'] = np.log(Year_data2['Close']) - np.log(Year_data2['Preclose'])

# 方法3：使用groupby按年分组
data_new4 = data_new.copy()
data_new4['year'] = data_new4.index.year
Year_data3 = data_new4.groupby('year')['Log_return'].sum().to_frame()
Year_data3['Raw_Return'] = np.exp(Year_data3['Log_return']) - 1

# 显示结果
print("年度对数收益率汇总:")
Year_data1
print("\n年末价格计算的收益率:")
Year_data2
print("\n使用groupby计算的年度收益率:")
Year_data3
```

```python
#计算滚动收益
## 基于对数收益率累加的滚动原始收益率
rolling_returns = pd.DataFrame()
windows = [5, 10, 20, 30, 60]
for window in windows:
    log_return_roll = data_new['Log_return'].rolling(window=window).sum()
    rolling_returns[f'Rolling_{window}d_Return'] = np.exp(log_return_roll) - 1

## 方法2：基于价格变化的滚动收益率（pct_change直接计算）
rolling_price_returns = pd.DataFrame()
for window in windows:
    # periods参数为“窗口长度-1”，即计算window天的价格变化（如5天滚动→比较今日与4天前的价格）
    rolling_price_returns[f'Rolling_{window}d_Price_Return'] = data_new['Close'].pct_change(periods=window-1)

## 展示结果（Jupyter环境用display更美观，非Jupyter可保留print）
print("滚动收益率（基于对数收益率累加）：")
display(rolling_returns.tail())  # 展示最新窗口的收益

print("\n滚动收益率（基于价格直接变化）：")
display(rolling_price_returns.tail())
```

```python
# 计算累积收益率
# 这是最准确的方法，特别是对于长期累积
cumulative_returns = pd.DataFrame()
cumulative_returns['Cumulative_Log_Return'] = data_new['Log_return'].cumsum()
cumulative_returns['Cumulative_Return'] = np.exp(cumulative_returns['Cumulative_Log_Return']) - 1

# 方法2：使用cumprod函数直接累乘(1+r)
# 这种方法在金融实践中也很常见
cumulative_returns['Cumulative_Return_Prod'] = (1 + data_new['Raw_return']).cumprod() - 1

# 方法3：使用pandas的累积函数
cumulative_returns['Cumulative_Return_Alt'] = data_new['Raw_return'].add(1).cumprod().sub(1)

# 显示结果
print("不同方法计算的累积收益率:")
cumulative_returns
```
